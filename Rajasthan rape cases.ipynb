```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_yearly = pd.read_csv('Rajasthan_rape_yearly.csv')
```

```python
df_district = pd.read_csv('Rajasthan_rape_district.csv')
```

plt.tight_layout() = Automatically adjusts plot parameters for a tight layout. This means it prevents labels, titles, and other plot elements from overlapping or being cut off, especially when you have multiple subplots or long labels.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_yearly, x='Year', y='Cases', marker='o', linewidth=2, color='crimson')
plt.title('Year-wise Rape Cases in Rajasthan (20182022)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.tight_layout()
plt.show()
```

 Insight: can visually spot increases or decreases over the years.

```python
#Plot bar chart of cases by district

plt.figure(figsize=(10, 6))
sns.barplot(data=df_district.sort_values('Cases_2022', ascending=False), x='Cases_2022', y='District', palette='Reds_r')
plt.title('Rape Cases by District in Rajasthan (2022)', fontsize=14)
plt.xlabel('Number of Cases')
plt.ylabel('District')
plt.tight_layout()
plt.show()
```

Insight: This shows which districts are the most affected in 2022.

```python
#Year with max cases
max_year = df_yearly[df_yearly['Cases'] == df_yearly['Cases'].max()]
print("Year with highest reported cases:\n", max_year)
```

```python
#District with max cases in 2022
max_dist = df_district[df_district['Cases_2022'] == df_district['Cases_2022'].max()]
print("District with highest cases in 2022:\n", max_dist)
```

#### 1. Which districts are reporting high cases?

### 2. Is the overall trend increasing or decreasing?
I checked the total yearly cases from 2018 to 2022. The numbers are going up and down, not a smooth trend.

- In 2018 it was 4335

- Jumped in 2019 to almost 6000

- Then dropped in 2020 (maybe due to lockdowns?)

- Again highest in 2021 with over 6300

- Then dropped a bit in 2022

So, its not clearly increasing or decreasing, but the level is overall high after 2019, so the issue is still serious.

### 3. Is there a big gap between top 5 districts and others?
Yes, definitely. Bhilwara has 301 cases and Ajmer has 206. Thats almost 100 cases difference. So the top 23 districts have much more cases than the rest.
